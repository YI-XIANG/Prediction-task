```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

train = pd.read_csv(r"./Data/Result.csv")
```

```python
True_Revenue = train[train['Revenue']==1]
```

```python
True_Revenue
```

```python
#Bar Data
def Plot_data(name):
    All_count = []
    Reve_count = []
    
    All_data = train[name].value_counts(sort='default')
    Reve_data = True_Revenue[name].value_counts(sort='default')
    sort_data = sorted(All_data.index)
    
    for i in sort_data:
        All_count.append(All_data[i])
        try:
            Reve_count.append(Reve_data[i])
        except:
            Reve_count.append(0)
       
    return sort_data, All_count, Reve_count

#plot 
def Plot_bar_charts(x1, y1, y2, Xlabel, Ylabel, Title):
    x_bar = np.arange(len(x1))
    width = 0.3
    plt.bar(x_bar, y1, width, color='cadetblue', label='All Data')
    plt.bar(x_bar + width, y2, width, color='deepskyblue', label='True(Revenue)')
    plt.xticks(x1)
    plt.ylabel(Xlabel)
    plt.ylabel(Ylabel)
    plt.title(Title)
    plt.legend(bbox_to_anchor=(1,1), loc='upper left')
    plt.show()
```

```python
x, y, z = Plot_data('Month')
Plot_bar_charts(x, y, z, "Month", "Count", "Month_Revenue")
```

```python
x, y, z = Plot_data('OperatingSystems')
Plot_bar_charts(x, y, z, "OperatingSystems", "Count", "OperatingSystems_Revenue")
```
