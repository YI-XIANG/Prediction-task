```python
Month_dt = train['Month'].value_counts(sort='default')
Month = sorted(Month_dt.index)
Month_count = []
for i in Month:
    Month_count.append(Month_dt[i])

plt.bar(Month,Month_count)
plt.xticks(Month, Month)
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Train Data')
plt.show()
```

```python
Test_dt = test['Month'].value_counts(sort='default')
Test = sorted(Test_dt.index)
Test_count = []
for i in Test:
    Test_count.append(Test_dt[i])

plt.bar(Test,Test_count)
plt.xticks(Test,Test)
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Test Data')
plt.show()
```
