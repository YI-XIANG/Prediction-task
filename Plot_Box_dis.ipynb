```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm

train = pd.read_csv(r"./Data/Result.csv")
```

```python
def Plot_box_dis(name):
    fig,axes=plt.subplots(1,2)
    fig.set_figheight(5)
    fig.set_figwidth(12)
    sns.boxplot(train[name],palette="Set3",ax=axes[0])
    sns.distplot(train[name],ax=axes[1])
```

```python
Plot_box_dis('Administrative')
```
